```python
from collections import Counter
from typing import Dict, List, Tuple, Union

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression, LinearRegression, Lasso
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from catboost import CatBoostClassifier
from sklearn.preprocessing import StandardScaler


```

```python
#загружаем таблицу
data = pd.read_csv('train.csv')
#data = data[data['Fare'] > 1.0]
data.head(5)
```

```python
data = data.drop(['Name', 'SibSp', 'Parch', 'Ticket','Cabin', 'Embarked','PassengerId'], axis=1)
data_ = data.dropna()
data_.head(5)
```

```python
for col in ['Sex']:
    data_ = pd.concat([data_, pd.get_dummies(data_[col])], axis=1)
    data_ = data_.drop(col, axis=1)
```

```python
y = data_['Survived']
data_ = data_.drop(['Survived'], axis=1)
data_.head(5)
```

```python
#data_ = StandardScaler().fit(data_).transform(data_)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_, y, test_size=0.2)
```

```python
# пойдем по порядку: логистическая регрессия 
model = LogisticRegression()#создание модели
model.fit(x_train, y_train) #обучение на тренировочном наборе
print(roc_auc_score(model.predict(x_test), y_test))# cчитаем метрику качества auc-roc 0.5- полное говно 1 - идеальная модель
```

```python
def predict_passenger(passenger_data):
    # преобразуем словарь в DataFrame
    passenger_df = pd.DataFrame([passenger_data])
    
    # масштабируем данные
    #passenger_df = StandardScaler().fit_transform(passenger_df)
    
    # делаем предсказание
    prediction = model.predict(passenger_df)
    
    return prediction


passenger_data = {
    'Pclass': 3,
    'Age': 20,
    'Fare': 10.0,
    'female': True,
    'male': False
}

prediction = predict_passenger(passenger_data)
print(prediction)
```
